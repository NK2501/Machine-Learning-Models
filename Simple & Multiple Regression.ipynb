```python
#Prediction on test dataset
y_pred=lr.predict(x_test.values.reshape(-1,1))
y_pred
```

```python
m=lr.coef_
m
```

```python
c=lr.intercept_
c
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
mse
```

```python
#Check for r2 score
r2=r2_score(y_test,y_pred)
r2
```

```python
#create a function to visualize the numerical columns
def histogram(column_name):
    sns.histplot(data=car[column_name],bins=20,kde=True)#kde is probabilitry density function
    plt.title(column_name+'distribution')
    plt.show()
```

```python
#create a function to visualize the categorical columns
def count_plot(column):
    plt.figure(figsize=(10,6))
    sns.countplot(data=car,x=column)
    plt.title(f"count plot:(column)")
    plt.xlabel(column)
    plt.ylabel("count")
    plt.xticks(rotation=90)
    plt.show()
```

```python
car['CarName'].unique()
```

```python
# Create a function
def clean_car_name(car_name):
    return car_name.split(" ")[0].lower()
```

```python
#apply the clean_car_name function to the canName column
car['CarName']=car['CarName'].apply(clean_car_name)
```

```python
# see the unique values after fixing
car['CarName'].unique()
```

```python
car['CarName']=car['CarName'].str.replace('vw','volkswagen')
car['CarName']=car['CarName'].str.replace('vokswagen','volkswagen')
car['CarName']=car['CarName'].str.replace('toyouta','toyota')
car['CarName']=car['CarName'].str.replace('maxda','mazda')
car['CarName']=car['CarName'].str.replace('porcshce','porsche')
count_plot('CarName')
```

# Bivariate Analysis

```python
# list each of categorical and Numerical columns

categorical_columns=car.select_dtypes(include=object).columns.tolist()
numerical_columns=car.select_dtypes(exclude=object).columns.tolist()


# create box plots for each categorical column


for column in categorical_columns:
        plt.figure(figsize=(10,6))
        sns.boxplot(x=column,y='price',data=car)
        plt.title(f'Box Plot:{column}vs.Price')
        plt.xlabel(column)
        plt.ylabel('Price')
        plt.xticks(rotation=90)
        plt.show()
```

```python
car
```

```python
correlation_matrix=car[numerical_columns].corr()
plt.figure(figsize=(16,10))
sns.heatmap(correlation_matrix,annot=True,cmap='Blues')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#list the dropped columns
dropped_columns=['car_ID','symboling']
car.drop(columns=dropped_columns,inplace=True)
numerical_columns.remove('car_ID')
numerical_columns.remove('symboling')
```

```python
# use lable encoder to encode the categorical columns
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
for column in categorical_columns:
    car[column]=label_encoder.fit_transform(car[column])
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
car[numerical_columns]=scaler.fit_transform(car[numerical_columns])
```

```python
# Build THE Model
# Lets divide the dataset into dependent and independent columns
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.3,random_state=23)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train.values.reshape(-1,1),y_train)#-1 Act as placeholder
y_pred=lr.predict(x_test.values.reshape(-1,1))
print(y_pred)
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_pred)
print("mse:",mse)
r2=r2_score(y_test,y_pred)
print("r2:",r2)
```
