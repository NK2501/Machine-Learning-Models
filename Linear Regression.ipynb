```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
ycap=2.76805098*x+5.590272428041182
plt.scatter(x,y)
plt.plot(x,ycap,c='blue')
plt.show()
```

```python
insure=pd.read_csv("Insure_auto.csv",index_col=0)
```

```python
ins_train=insure.iloc[[1,4,5,7,2,9,3],:]
ins_test=insure.iloc[[6,0,8],:]
x_train=ins_train[['Home','Automobile']].values
y_train=ins_train[['Operating_Cost']].values
x_test=ins_test[['Home','Automobile']].values
y_test=ins_test[['Operating_Cost']].values

print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)

lr= LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
from sklearn.model_selection import train_test_split
```

```python
#ins_train=insure.iloc[[1,4,5,7,2,9,3],:]
#ins_test=insure.iloc[[6,0,8],:]
ins_train,ins_test=train_test_split(insure,test_size=0.3,random_state=23)
x_train=ins_train[['Home','Automobile']].values
y_train=ins_train[['Operating_Cost']].values
x_test=ins_test[['Home','Automobile']].values
y_test=ins_test[['Operating_Cost']].values

print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)

lr= LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print(r2_score(y_test,y_pred))
```
